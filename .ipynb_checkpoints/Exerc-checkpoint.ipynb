```python
import tabula as tb

# Extraindo a tabela da página 2 do PDF
tabelas = tb.read_pdf("Modelo_PDF (1).pdf", pages="all", lattice=True)

# Substituir \r por espaço vazio nos cabeçalhos das tabelas e no corpo da tabela
for i, tabela in tabelas:
    tabela.columns = [col.replace("\r", " ") for col in tabela.columns]
    tabela = tabela.applymap(lambda x: x.replace("\r", " ") if isinstance(x, str) else x)
    tabelas[i] = tabela.fillna("")
    
# Exibir as tabelas
for i, tabela in enumerate(tabelas):
    display(tabela)
```
